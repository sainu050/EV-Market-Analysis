## Exploratory Data Analysis and Visualization

1. Year-wise EV Sales Growth

```python
yearly_sales = sales.iloc[:, 2:].sum()

yearly_sales.plot(kind='line', marker='o')
plt.title("Year-wise EV Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

Insight:
The graph shows a sharp rise in EV sales after 2020, indicating accelerated market adoption. The growth can be attributed to increasing consumer awareness, government incentives, and improvements in EV technology and charging infrastructure.

2. Top 10 EV Manufacturers

```python
sales["Total"] = sales.iloc[:, 2:].sum(axis=1)

top_makers = sales.groupby("Maker")["Total"].sum().sort_values(ascending=False).head(10)

top_makers.plot(kind='bar')
plt.title("Top 10 EV Manufacturers")
plt.ylabel("Total Sales")
plt.xticks(rotation=85)
plt.show()
```

Insight:
The analysis reveals that Ola Electric Technologies Pvt Ltd is the leading EV manufacturer in the dataset, indicating its strong market presence and significant contribution to India's EV industry. This suggests that the Indian EV market is currently concentrated among a few major players with strong brand presence and production capacity.

3. Vehicle Category Analysis

```python
cat_sales = sales.groupby("Cat")["Total"].sum()

cat_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("EV Sales by Vehicle Category")
plt.ylabel("")
plt.show()
```

Insight:
Two-wheelers represent the largest share of EV sales, indicating that affordability, lower operating costs, and suitability for daily commuting have made electric scooters and bikes the most popular EV segment.

4. Charging Stations by State

```python
top_states = charging.sort_values("No. of Operational PCS",ascending=False).head(10)

sns.barplot(x="No. of Operational PCS",y="State",data=top_states)

plt.title("Top States by Charging Stations")
plt.show()
```

Insight:
Maharashtra has the highest number of operational charging stations, indicating strong EV infrastructure development. The availability of charging facilities can play an important role in encouraging EV adoption and reducing range anxiety among users.

5. EV Makers by State

```python
state_count = maker_place["State"].value_counts().head(10)

state_count.plot(kind='bar')
plt.title("States with Most EV Makers")
plt.ylabel("Number of Makers")
plt.show()
```

Insight:
Maharashtra has the highest number of EV manufacturers among all states in the dataset. This indicates that the state has developed a strong EV ecosystem supported by industrial infrastructure, market opportunities, and favorable conditions for EV businesses.

6. Most Popular Vehicle Class

```python
top_classes = vehicle_class.sort_values(
    'Total Registration',
    ascending=False
)

sns.barplot(
    x='Total Registration',
    y='Vehicle Class',
    data=top_classes
)

plt.title("Most Popular EV Vehicle Classes")
plt.xlabel("Total Registrations")
plt.ylabel("Vehicle Class")

plt.show()
```

Insight:
The vehicle class analysis reveals that two-wheelers have the highest number of registrations, making them the most popular EV vehicle class. Their lower purchase cost, better energy efficiency, and convenience for urban transportation have contributed significantly to their widespread adoption.

7. Category-wise Growth Over Years

```python
cat_year = sales.groupby('Cat').sum(numeric_only=True)
cat_year = cat_year.drop(columns='Total') #dataframe already contains a Total column, it will also appear on the x-axis
cat_year = cat_year.T #Currently, years are columns. For plotting, years should become rows.
```

```python
plt.figure(figsize=(10,5))
for category in cat_year.columns:
    plt.plot(
        cat_year.index,
        cat_year[category],
        marker='o',
        label=category
    )

plt.title("Category-wise EV Growth Over Years")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.legend(title="Category")
plt.grid(True)

plt.show()
```

Insight:
The graph indicates that two-wheelers have experienced the fastest growth over the years, contributing significantly to overall EV adoption. Three-wheelers and four-wheelers have also shown steady growth, indicating expanding acceptance across different transportation sectors.
